```python
Math = int(input("Enter the Math score: "))
Science = int(input("Enter the Science score: "))
English = int(input("Enter the English score: "))

average = (Math + Science + English) / 3

if average = 100:
    print("Waouh you got the Excellent Score!")
elif average >= 90:
    print("Excellent! You got an A.")
elif average >= 80:
    print("Very Good! You got a B.")
elif average >= 70:
    print("Good! You got a C.")
elif average >= 60:
    print("Needs Improvement! You got a D.")
elif average < 0:
    print("Fail! You got an F.")
else:
    print("Invalid Input. Try another number")


```
