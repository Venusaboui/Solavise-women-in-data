```python
Math = int(input("Enter the Maths mark: "))
Science = int(input("Enter the Science mark: "))
English = int(input("Enter the English mark: "))

average = (Math + Science + English) / 3
print(" The average is:", average)

if average == 100:
    print("Waouh you got the Excellent Score!")
elif 90 <= average <= 99:
    print("Excellent! You got an A")
elif 80 <= average <= 89:
    print("Very Good! You got a B")
elif 70 <= average <= 79:
    print("Good! You got a C")
elif 60 <= average <= 69:
    print("Needs Improvement! You got a D")
elif average < 60 :
    print("Fail! You got an F")
elif average == 0 :
    print(" Oh no! You failed everything")
else :
    print("Invalid Input. Try again!")
```
